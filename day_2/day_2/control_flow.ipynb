## 1. Function Scope

In Python, variables defined inside a function are **local** to that function.  

Variables defined outside any function are **global**. When a function is called, Python creates a new **stack frame** -- a separate namespace where local variables live.

The diagram below shows what happens in memory when `adder()` is called:

```
Global Frame        adder() Stack Frame
x = 100             x = 10   (local, shadows global)
y = 200             y = 20   (local, shadows global)
                    result = 30
```

````python
# --- Global scope ---
x = 100
y = 200

print(f"[Global] : x = {x}, y = {y}")


def adder():
    # --- Local scope (new stack frame when called) ---
    x = 10
    y = 20
    result = x + y
    print(f"[Inside adder] : x = {x}, y = {y}, result = {result}")
    return result


total = adder()

# Back in global scope -- x and y are unchanged
print(f"[Global after call]  x = {x}, y = {y}")
print(f"[Global] total returned by adder = {total}")
````

## 2. Conditionals

A **conditional** lets your program choose which code to run based on whether a
condition is `True` or `False`. Python provides three branching keywords:

| Keyword | When it runs |
|---------|-------------|
| `if`    | Condition is `True` |
| `elif`  | Previous conditions were `False`, this one is `True` |
| `else`  | None of the above conditions were `True` |

Rules:
- Every block starts with exactly one `if`.
- `elif` and `else` are optional; you can have as many `elif` as needed.
- `else` has no condition — it's a catch-all.

### 2a. `if` only — one branch

````python
temperature = 35

if temperature > 30:
    print("It's hot outside.")

print("Done.")   # always runs
````

### 2b. `if / else` — two mutually exclusive branches

````python
temperature = 20

if temperature > 30:
    print("It's hot outside.")
else:
    print("It's not hot outside.")

print("Done.")
````

### 2c. `if / elif / else` — multiple exclusive branches

Use `elif` when you have more than two distinct cases.
Only the **first** branch whose condition is `True` executes.

````python
score = 72

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score {score} → Grade {grade}")
````

### 2e. Defensive Guard Pattern — Early Return on Bad Data

A **guard clause** is an `if` at the top of a function that checks for
invalid input and returns (or raises an error) immediately, *before* any
real work is done.

Benefits:
- Keeps the "happy path" unindented and easy to read.
- Fails fast with a clear error message instead of a cryptic crash later.

````python
def calculate_average(numbers):
    # --- Guards ---
    if not isinstance(numbers, list):
        print("Error: input must be a list.")
        return None
    
    if len(numbers) == 0:
        print("Error: list is empty.")
        return None

    # --- Happy path (only reached if guards pass) ---
    total = sum(numbers)
    average = total / len(numbers)
    return average


# Test all three paths
print(calculate_average([10, 20, 30]))   # happy path  → 20.0
print(calculate_average([]))             # empty list  → None
print(calculate_average("oops"))         # wrong type  → None
````

## 3. For Loops & Iterables

A **`for` loop** repeats a block of code once for each item in a sequence.
The sequence it loops over is called an **iterable** — any object that can
hand out its items one at a time.

```python
for <variable> in <iterable>:
    # body runs once per item
```

The only iterable we've seen so far is the **list**. Each pass through the
loop, Python binds the next list item to `<variable>` automatically.

### 3a. Basic `for` loop over a list

````python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])
# string slicing will return a string, unlike with the above which returns an element of the list...
print(fruits[:1])

for fruit in fruits:
    print(fruit)
````

### 3b. Doing work inside a loop — accumulation

A common pattern is to start with a running total (or empty container)
outside the loop, then update it on each iteration.

````python
prices = [4.99, 12.50, 3.75, 8.00]

total = 0
for price in prices:
    total = total + price

print(f"Total: ${total:.2f}")   # → Tota1: $29.24

# len() is a built-in function that returns the number of items in a list (or characters in a string, etc.) .;/
print(total / len(prices))   # → 7.31

#apply this similar pattern for the TLAB
````

### 3c. Iterables vs Iterators

Two related but distinct concepts:

| Term | What it is | Key method |
|------|-----------|------------|
| **Iterable** | Any object you can loop over (e.g. a list) | `__iter__()` — returns an iterator |
| **Iterator** | The "cursor" object that tracks the current position | `__next__()` — returns the next item |

- `iter(x)` converts an iterable `x` into an iterator.
- `next(it)` pulls the **next** item from iterator `it`.
- When items run out, `next()` raises a `StopIteration` exception.

````python
colors = ["red", "green", "blue"]

# Convert the list (iterable) into an iterator
it = iter(colors)

print(next(it))   # red
print(next(it))   # green
print(next(it))   # blue

# Uncomment the line below to see StopIteration:
# print(next(it))
````

### 3d. The Map Pattern — Transforming a List

A **map** operation takes every item in a list, applies a transformation to it,
and collects the results into a **new** list.

The recipe always looks the same:
1. Create an empty result list **before** the loop.
2. Loop over the **source** list.
3. Transform each item and `append` to the result list.

````python
#clean_heartrate_data uses this same pattern to loop over the raw data and build a new list of cleaned values
words = ["hello", "world", "python"]

# Map: lowercase -> UPPERCASE
uppercased = []                         # 1. empty result list
for word in words:                      # 2. loop over SOURCE list
    uppercased.append(word.upper())     # 3. transform + collect

print(uppercased)   # ['HELLO', 'WORLD', 'PYTHON']
print(words)        # ['hello', 'world', 'python']  <- unchanged
````

### 3e. Strings Are Immutable

Calling `.upper()` does **not** modify the original string — it returns a
brand-new string. Every string method that "transforms" actually produces a
new object; the original is always left untouched.

````python
name = "alice"
name.upper()          # returns "ALICE" but result is discarded
print(name)           # still "alice"

# To keep the result, assign it:
loud_name = name.upper()
print(loud_name)      # "ALICE"
print(name)           # "alice" -- original still untouched
````

### 3f. Danger: Never Append to the List You're Iterating Over

If you loop over a list and `.append()` to that **same** list inside the
body, the list grows every iteration and the loop never ends -> **infinite loop**.

> ⚠️ The broken code below is commented out — do **not** uncomment and run it.

### 4b. Stripping Newline Characters — `.strip()` and `.rstrip()`

Each line from `.readlines()` ends with `\n`. Before you can use the value
as a number you need to remove that trailing character.

| Method | Removes from | Example |
|--------|--------------|---------|
| `.strip()`  | both ends | `"  hi\n".strip()` → `"hi"` |
| `.rstrip()` | right end only | `"4.99\n".rstrip()` → `"4.99"` |
| `.lstrip()` | left end only  | `"  hi".lstrip()` → `"hi"` |

> **Recall (cell 3f):** string methods return a **new** string — the original
> is never modified.

````python
raw_line = "4.99\n"           # what .readlines() gives you
print(f"Before strip: {repr(raw_line)}")

clean_line = raw_line.strip()
print(f"After strip:  {repr(clean_line)}")
print(f"Original raw_line unchanged: {repr(raw_line)}")   # still has \n
````

### 4c. Typecasting — `int()`, `float()`, `str()`

Even after stripping, the value is still a **string**. To do math you
must cast it to a numeric type.

| Function | Converts to | Example |
|----------|------------|--------|
| `int(x)`   | integer | `int("42")` → `42` |
| `float(x)` | float   | `float("4.99")` → `4.99` |
| `str(x)`   | string  | `str(99)` → `"99"` |

> **Note:** `float()` tolerates surrounding whitespace (`float("4.99\n")` works),
> but calling `.strip()` first makes your intent explicit and readable.

````python
raw = "4.99\n"

value_a = float(raw.strip())
print(f"float('4.99'): {value_a}  type: {type(value_a)}")

discounted = value_a * 0.9
print(f"10% discount:  {discounted:.2f}")

text_value = "4.99"
# print(text_value + 1)   # <-- TypeError: can only concatenate str to str

numeric_value = float(text_value)
print(f"numeric_value + 1 = {numeric_value + 1}")
````
