```python
# TODO: import all the packages that you will need for this notebook here
import statistics as stats
import matplotlib.pyplot as plt

import cleaner
```

```python
# TODO: read your text files into your code using file I/O
control_file = open("data/control.txt")
treatment_file = open("data/treatment.txt")

control_data = control_file.readlines()
treatment_data = treatment_file.readlines()

# TODO: clean both list objects that you created above using the 'clean_data' function
clean_control = cleaner.clean_data(control_data)
clean_treatment = cleaner.clean_data(treatment_data)

print("Clean Control:", clean_control)
print("Clean Treatment:", clean_treatment)
```

## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group

std_control = round(stats.stdev(clean_control), 2)
std_treatment = round(stats.stdev(clean_treatment), 2)

# TODO: print your findings
print("Standard Deviation for Control:",std_control,
      '\n'"Standard Deviation for Treatment:", std_treatment)
```

```python
# TODO: Calculate the mean of both the control and treatment group
mean_control = round(stats.mean(clean_control), 2)
mean_treatment = round(stats.mean(clean_treatment), 2)

# TODO: print your findings
print("Mean for Control:", mean_control,
      '\n'"Mean for Treatment:", mean_treatment)
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []

for outliers in clean_control:
    if outliers < mean_control -2*std_control or outliers > mean_control + 2*std_control:
        outliers_control.append(outliers)

# how many outliers? use length()
print("Control Outlier Count:", len(outliers_control), '\n' "Control Outliers:", outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group? (use 2 * stdev to find these outliers)
outliers_treatment = []

for outliers in clean_treatment:
    if outliers < mean_treatment -2*std_treatment or outliers > mean_treatment + 2*std_treatment:
        outliers_treatment.append(outliers)

print("Treatment Outlier Count:", len(outliers_treatment), '\n' "Treatment Outliers:", outliers_treatment)
```

## 3. Create Data Visualizations

For this final step, generate data visualizations to view the dispersion of both of your groups.

Make sure that the two visualizations you create are on the same figure. 

Make a histogram

```python
# TODO: create a data visualization to view differences in dispersion and mean

fig, ax = plt.subplots()

ax.set_title("Control vs Treatment on Workshop Viewing Times")
ax.set_xlabel("Minutes")
ax.set_ylabel("Employees")

# create the histogram for both the cleaned control group and cleaned treatment group
ax.hist(clean_control, bins = 25, color = 'hotpink', edgecolor ='magenta', alpha = 0.6, label = 'Control')
ax.hist(clean_treatment, bins = 25, color = 'cyan', edgecolor ='blue', alpha = 0.5, label = 'Treatment')

# add a legend for clarity 
ax.legend()
```
