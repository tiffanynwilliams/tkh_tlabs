## 1. Probability Refresher

### Elementary Event
An **elementary event** is the simplest possible outcome of a random experiment â€” a single, indivisible result. It cannot be broken down into smaller outcomes.

- Flipping a coin and getting **Heads** is an elementary event.
- Rolling a die and getting **4** is an elementary event.

---

### Non-Elementary Event
A **non-elementary event** (also called a *compound event*) is any outcome made up of **two or more elementary events** combined.

- Rolling a die and getting an **even number** is a non-elementary event â€” it includes the outcomes {2, 4, 6}.
- Drawing a **face card** from a deck is non-elementary â€” it includes Jacks, Queens, and Kings across all suits.

---

### Frequentist Statistics
**Frequentist statistics** defines probability based on **long-run frequency** â€” how often an event occurs over a large number of repeated trials. If you flip a fair coin 10,000 times, it will land Heads close to 50% of the time. Probability is treated as a fixed, objective fact about the world.

---

### Bayesian Statistics
**Bayesian statistics** defines probability as a **degree of belief** that is updated as new evidence is collected. You start with a prior belief (e.g., "I think this coin is fair"), observe data, and update your belief accordingly. Probability is treated as subjective and dynamic.

---

## 2. Probability Distributions

A **probability distribution** describes how probability is spread across all possible outcomes of a random experiment. Instead of looking at a single event in isolation, a distribution gives you a complete picture â€” every outcome and how likely each one is.

A valid probability distribution has two properties:
- Every probability is between **0** and **1**
- All probabilities **sum to 1** (something must happen)

A **bar graph** is one of the simplest ways to visualize a probability distribution. Each bar represents one outcome, and its height represents how probable that outcome is.


````python
import matplotlib.pyplot as plt

pet_probs = {
    "dog":     0.5,
    "cat":     0.3,
    "hamster": 0.1,
    "lizard":  0.0,
    "no pet":  0.1
}

# --- Pet Ownership Distribution ---
pet_labels = list(pet_probs.keys())
pet_values = list(pet_probs.values())

print(pet_labels)
print(pet_values)
````

````python
plt.figure()
plt.bar(pet_labels, pet_values)

plt.title("Household Pet Ownership â€” Probability Distribution")
plt.xlabel("Pet")
plt.ylabel("Probability")
plt.ylim(0, 1)
````

## 4. Types of Probability Distributions

### Starting from the Bottom Up

In the Zillow example above, we worked with **real data**.
We counted prices, divided by the total, and got probabilities directly from observations.

But in the real world, you often do **not** have all the data.
You might have a small sample, or no data at all yet.
In those situations, mathematicians noticed that many real-world phenomena
tend to follow **recurring patterns**: shapes that show up again and again
across very different domains.

These patterns were given names and studied carefully.
We call them **probability distribution models**.

> **Important:** These models are not perfect descriptions of reality.
> No real dataset is a perfect normal curve or a perfect uniform distribution.
> They are **simplified tools** that help us reason, estimate, and make decisions
> when we cannot count every single data point.
> Think of them the way you think of a map, a map is not the territory,
> but it is still incredibly useful for navigation.

The three most common distribution models you will encounter are:

| Distribution | Shape | Best for |
|---|---|---|
| **Normal** | Bell curve, symmetric | Heights, test scores, measurement errors |
| **Binomial** | Discrete bars | Counting successes in repeated yes/no trials |
| **Uniform** | Flat, equal bars | Random picks where every outcome is equally likely |

---

### The Normal Distribution

The **normal distribution** (also called the *bell curve*) is the most famous distribution in statistics.
It describes data that clusters symmetrically around a central value.

**Key characteristics:**
- Perfectly **symmetric** — the left half mirrors the right half
- The **mean**, **median**, and **mode** are all the same value (the center)
- Most values fall **close to the mean**; extreme values are rare
- Described by two numbers: the **mean** (where it is centered) and the **standard deviation** (how spread out it is)

**Real-world examples where data is approximately normal:**
- Heights of adults in a population
- SAT scores across thousands of students
- Measurement errors in a manufacturing process
- Blood pressure readings in a healthy population

**The 68-95-99.7 rule** (a practical shortcut):
- ~68% of values fall within **1 standard deviation** of the mean
- ~95% of values fall within **2 standard deviations** of the mean
- ~99.7% of values fall within **3 standard deviations** of the mean

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random heights (in inches) following a normal distribution
# mean = 67 inches (~5 ft 7 in),  std = 3 inches
mean   = 67
std    = 3

heights = np.random.normal(mean, std, 1000)

print(heights)
````

````python
# Plot the histogram
plt.figure()
plt.hist(heights)

# Mark the mean with a vertical line
plt.axvline(mean, color='red', linewidth=2, label='mean')

# Mark +/- 1 standard deviation
plt.axvline(mean + std, color='orange', linewidth=1, linestyle='--', label='+1 std')
plt.axvline(mean - std, color='orange', linewidth=1, linestyle='--', label='-1 std')

# TODO: title this graph "Normal Distribution — Adult Heights"
...

plt.show()
````

````python
# --- Practical probability question ---
# What fraction of people are taller than 70 inches?

taller_than_70 = 0

for h in heights:
    if h > 70:
        taller_than_70 = taller_than_70 + 1

prob_tall = taller_than_70 / len(heights)

print('P(height > 70 inches): ', round(prob_tall, 3))
````

### The Binomial Distribution

The **binomial distribution** answers a very specific question:

> *If I repeat a yes/no experiment N times, how likely am I to get exactly K successes?*

Each individual trial must have exactly two outcomes: **success** or **failure**.
Each trial must be **independent** (the result of one flip does not affect the next).

**Key characteristics:**
- Works with **discrete** counts (you cannot flip a coin 2.5 times)
- Defined by two numbers: **n** (number of trials) and **p** (probability of success per trial)
- When p = 0.5 and n is large, it looks like a normal distribution
- When p is close to 0 or 1, it is skewed

**Real-world examples:**
- Flipping a coin 20 times — how often do you get exactly 10 heads?
- A doctor treats 50 patients — how many recover if each has a 70% recovery rate?
- A company sends 100 emails — how many are opened if the open rate is 25%?

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulate flipping a coin 20 times, repeated 1000 times
# Record how many heads appear in each set of 20 flips

n = 20       # number of flips per trial
p = 0.5      # probability of heads
runs = 1000  # number of times we repeat the 20-flip experiment

results = np.random.binomial(n, p, runs)

print(results)
````

````python
# Plot the distribution of outcomes
plt.figure()
plt.hist(results)

plt.title('Binomial Distribution — Heads in 20 Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Count (out of 1000 experiments)')

plt.show()
````

````python
# --- Practical probability question ---
# What fraction of experiments produced exactly 10 heads?

exactly_10 = 0

for result in results:
    if result == 10:
        exactly_10 = exactly_10 + 1

prob_10 = exactly_10 / runs

print('P(exactly 10 heads in 20 flips): ', round(prob_10, 3))
````

### The Uniform Distribution

The **uniform distribution** is the simplest of all distributions.
Every outcome has the **exact same probability** of occurring.
There is no clustering, no peak, no skew — just a perfectly flat line.

**Key characteristics:**
- All outcomes are **equally likely**
- Defined by a **minimum** and a **maximum** value
- The histogram looks like a flat rectangle
- The mean sits exactly in the middle of the range

**Real-world examples (approximately uniform):**
- Rolling a fair six-sided die — each face has a 1/6 chance
- A random number generator picking any number between 0 and 1
- A bus that arrives at a random time within a 30-minute window
- Lottery ball draws

> **Note on real-world fit:** Perfect uniformity is rare in nature.
> But it is a useful starting assumption when you have **no reason to prefer one outcome over another**.

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulate rolling a fair 6-sided die 1000 times

rolls = np.random.randint(1, 7, size=1000)
# Count how often each face appeared
face_counts = []

for face in range(1, 7):
    count = 0
    for roll in rolls:
        if roll == face:
            count = count + 1
    face_counts.append(count)

print(face_counts)
````

````python
# Plot the distribution
plt.figure()
plt.bar([1, 2, 3, 4, 5, 6], face_counts, color='salmon', edgecolor='white')

# Draw a horizontal line at the expected count (1000 / 6)
expected = 1000 / 6
plt.axhline(expected, color='red', linewidth=1.5, linestyle='--', label='expected (equal)')

plt.title('Uniform Distribution — 1000 Dice Rolls')
plt.xlabel('Die Face')
plt.ylabel('Count')
plt.legend()

plt.show()
````

````python
# --- Practical probability question ---
# What is the probability of rolling a 4?

count_4 = 0

for roll in rolls:
    if roll == 4:
        count_4 = count_4 + 1

prob_4 = count_4 / len(rolls)

print('Observed P(roll = 4): ', round(prob_4, 3))
print('Expected P(roll = 4): ', round(1/6, 3))
````

### Key Takeaways

| Distribution | Shape | Defined by | Use when |
|---|---|---|---|
| **Normal** | Symmetric bell curve | mean, standard deviation | Data clusters around a center value |
| **Binomial** | Discrete bars, may be skewed | n (trials), p (success rate) | Counting successes in yes/no experiments |
| **Uniform** | Flat rectangle | min, max | Every outcome is equally likely |

**The big picture:**

1. Start with **real data** whenever you have it (like the Zillow example).
2. When data is limited, choose a **distribution model** whose shape matches what you know about the process.
3. Use the model to **estimate probabilities**, make predictions, and reason about outcomes you have not yet observed.
4. Always remember: the model is a tool, not the truth.
   As the statistician George Box put it — *all models are wrong, but some are useful*.

## 5. The Central Limit Theorem

### Building Up to It from Scratch

Imagine you are trying to find the **average rent** in a city of one million apartments.
You cannot survey every apartment — that would take years.
So instead, you pick a random sample of 50 apartments, calculate the average of that sample,
and use it to estimate the true city-wide average.

But here is a question: how reliable is that one sample average?
What if you had picked a different 50 apartments?
You would get a slightly different number.

Now imagine doing this **thousands of times** — pick 50 apartments, calculate the average, write it down, repeat.
You would end up with thousands of sample averages.
What shape would those averages form?

This is exactly what the **Central Limit Theorem** answers.

---

### The Central Limit Theorem (CLT)

> **If you take many random samples from any distribution and calculate the average of each sample,
> those averages will form a normal distribution — regardless of the shape of the original data.**

That last part is what makes it remarkable: **regardless of the shape of the original data.**

- The original data can be skewed, flat, bimodal, or any irregular shape.
- The rent prices in the Zillow example are right-skewed (a few very expensive outliers pull the tail right).
- Dice rolls follow a uniform distribution.
- None of these are normal.

But if you repeatedly sample from any of them and average each sample,
those averages will pile up into a bell curve.

**Two conditions for this to work:**
1. Each sample must be picked **randomly and independently**
2. The sample size must be **large enough** (a common rule of thumb is n >= 30)

---

### Why Does This Matter?

The CLT is one of the most important ideas in all of statistics because it is the foundation
for nearly every practical tool used in data analysis:

- **Polls and surveys** — A political poll samples ~1,000 people from millions of voters.
  The CLT tells us that the average opinion of that sample is normally distributed around the true population opinion,
  which lets us calculate a margin of error.

- **Quality control** — A factory samples 50 bolts per hour to check diameter.
  The CLT lets engineers use normal distribution math to catch when the machine drifts out of spec.

- **A/B testing** — A company tests two versions of a website.
  The CLT is what justifies using a normal-based test to decide if the difference in click rates is real.

- **Medical trials** — Researchers measure the average effect of a drug on a sample of patients.
  The CLT is why they can draw conclusions about the whole population from that sample.

Without the CLT, none of these tools would work unless the underlying data happened to be perfectly normal.
The CLT frees us from that requirement.

### How Sample Size Affects the Result

The CLT kicks in more strongly as the sample size grows.
With a **tiny sample** (n = 2 or 3), the averages still look irregular.
With a **larger sample** (n = 30+), the bell curve shape becomes clear.

There is also another pattern: as sample size increases,
the bell curve gets **narrower**.
This makes sense — a larger sample is a better estimate of the true average,
so there is less variability between samples.

The code below plots sample averages for three different sample sizes so you can see this directly.

````python
import random
import matplotlib.pyplot as plt

# Compare three sample sizes
sample_sizes = [2, 10, 30]
colors       = ['salmon', 'mediumseagreen', 'steelblue']
num_samples  = 1000

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for i in range(3):

    n = sample_sizes[i]
    averages = []

    for j in range(num_samples):
        sample = random.choices(rent_prices, k=n)

        total = 0
        for price in sample:
            total = total + price

        averages.append(total / n)

    axes[i].hist(averages, bins=30, color=colors[i], edgecolor='white')
    axes[i].set_title('Sample size = ' + str(n))
    axes[i].set_xlabel('Sample Average ($)')
    axes[i].set_ylabel('Count')

plt.suptitle('CLT: How Sample Size Shapes the Distribution of Averages', fontsize=12)
plt.tight_layout()
plt.show()

print('Larger samples produce a tighter, more normal bell curve.')
````

### Why the Central Limit Theorem Is So Important for Data Analysis

The CLT is often called the **backbone of inferential statistics** — the branch of statistics
that lets you draw conclusions about a large population from a small sample.

Here is why it changes everything:

**1. You do not need to know the shape of the original data.**
Income distributions, website load times, disease rates — these are all skewed or irregular.
The CLT says: it does not matter. Sample averages will behave normally anyway.

**2. It lets you quantify uncertainty.**
Because sample averages follow a normal distribution, you can use normal distribution math
to calculate exactly how confident you are in your estimate.
This is where **confidence intervals** and **margins of error** come from.

**3. It is the reason small samples are useful at all.**
Without the CLT, you would need to survey millions of people to say anything reliable.
With it, a well-designed sample of a few hundred is often enough.

**4. It underpins hypothesis testing.**
When a data analyst asks "is this result statistically significant?",
the math behind that answer almost always relies on the CLT.

> **One-sentence summary:**
> The Central Limit Theorem is why we can trust sample averages to represent populations,
> no matter what the underlying data looks like.

#### Exercise 4

The data below represents the **heights (in inches) of 200 adults**.
Run the code, look at the histogram, then answer the questions.

````python
import numpy as np
import matplotlib.pyplot as plt

# Heights of 200 adults — normally distributed
heights = list(np.random.normal(loc=67, scale=3.5, size=200).round(1))

# Calculate mean and standard deviation
total = 0
for h in heights:
    total = total + h

mean_height = total / len(heights)

total_sq_diff = 0
for h in heights:
    total_sq_diff = total_sq_diff + (h - mean_height) ** 2

std_height = (total_sq_diff / len(heights)) ** 0.5

# Plot
plt.figure()
plt.hist(heights, bins=20, color='steelblue', edgecolor='white')
plt.axvline(mean_height, color='red',    linewidth=2, label='mean')
plt.axvline(mean_height + std_height, color='orange', linewidth=1, linestyle='--', label='+1 std')
plt.axvline(mean_height - std_height, color='orange', linewidth=1, linestyle='--', label='-1 std')
plt.title('Adult Heights')
plt.xlabel('Height (inches)')
plt.ylabel('Count')
plt.legend()
plt.show()

print('Mean:               ', round(mean_height, 1))
print('Standard deviation: ', round(std_height, 1))
print('1 std range:        ', round(mean_height - std_height, 1), 'to', round(mean_height + std_height, 1))
print('2 std range:        ', round(mean_height - 2 * std_height, 1), 'to', round(mean_height + 2 * std_height, 1))
````
