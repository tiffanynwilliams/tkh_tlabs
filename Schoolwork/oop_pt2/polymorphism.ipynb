## 1. How do we use Objects when Designing Systems?

When building real-world software systems, we rarely start from scratch every time. Two of the most powerful tools in object-oriented design are **inheritance** and **polymorphism** — both of which help us write less code, reuse what we already have, and build systems that are easier to maintain.

---

### Inheritance

**Inheritance** allows one class (the *child* or *subclass*) to take on the attributes and methods of another class (the *parent* or *superclass*).

Think of it like an org chart at a company. Every employee shares some common traits — they have a name, a department, and a start date. But a Manager has additional responsibilities that a general Employee does not. Instead of rewriting all the shared logic, a `Manager` class can **inherit** from an `Employee` class and only add what is unique to managers.

**Why it matters:**
- Eliminates duplicate code across related classes
- Makes your codebase easier to extend — add a new role by subclassing, not rewriting
- Enforces a consistent structure across related objects

---

### Polymorphism

**Polymorphism** means "many forms." In Python, it refers to the ability of different classes to define the same method name, each with their own implementation.

Imagine a company platform that generates reports. A `SalesReport` and an `OperationsReport` are different, but both have a `generate()` method. Code that calls `.generate()` does not need to know which type of report it is working with — it just calls the method and gets the right result. That flexibility is polymorphism.

**Why it matters:**
- Lets you write general-purpose code that works across many object types
- Makes systems easier to extend — add a new class without changing existing logic
- Supports real-world design patterns used in production software

---

### Reusability in Practice

Together, inheritance and polymorphism are the foundation of **reusable system design**:

| Concept | What it solves | Real-world analogy |
|---|---|---|
| Inheritance | Shared structure across related objects | Job roles in a company hierarchy |
| Polymorphism | Consistent interface with different behaviors | Different report types, same `.generate()` call |

## 2. Inheritance — Building on What Already Exists

Now that we have a `Fellow` class, let's talk about **inheritance** and why it matters in system design.

### The Core Idea

Inheritance lets you define a new class that **automatically gets all the attributes and methods of an existing class**, without rewriting them. The existing class is called the **parent** (or base) class. The new class is called the **child** (or subclass).

In workforce development, not every fellow is the same. TKH runs multiple fellowship programs, each with its own focus, requirements, and outcomes. Rather than building a completely separate class for each program from scratch, we can have each program's class **inherit** from `Fellow` — keeping the shared structure and only adding what is unique.

---

### How `__init__` Works with Inheritance

When a child class defines its own `__init__`, it needs to handle two things:

**1. Call the parent's `__init__` first**

This is done with `super().__init__(...)`, which runs the parent class's setup and assigns the shared attributes (`fname`, `lname`, `track`, `tools`). Without this call, those attributes would never be set on the object.

**2. Then set the child's unique attributes**

After the parent is initialized, the child's `__init__` adds whatever extra attributes are specific to that subclass.

Think of it like onboarding at a company. Every new hire fills out the same paperwork (the parent `__init__`). But someone joining as a contractor also fills out additional forms specific to their role (the child's extra steps).

---

### The Two Subclasses We Will Build

| Class | Stands For | What makes it unique |
|---|---|---|
| `IFFellow` | Innovation Fellowship Fellow | Enrolled in a tech training cohort for NY & NJ; has a cohort number. |
| `TFFellow` | Technology Fellowship Fellow | Enrolled in a tech training cohort over in Stamford, CT; has a cohort number. |

Both inherit from `Fellow`, so they automatically have `fname`, `lname`, `track`, `tools`, and `switch_track()`. Each adds its own attributes on top.

Run the cells below to define each subclass.

---

```python
class IFFellow(Fellow):

    def __init__(self, fname, lname, track, tools, cohort_number, nonprofit_partner):

        # Step 1: Run Fellow's __init__ to set fname, lname, track, and tools
        # super is the parent class, it's like self...but for referring to the parent class
        super().__init__(fname, lname, track, tools)

        # Step 2: Set the attributes that are unique to IFFellow
        self.cohort_number      = cohort_number
        self.nonprofit_partner   = nonprofit_partner


# --- Test it out ---
if_fellow = IFFellow("Marcus", "Webb", "Data Science", ["Python", "SQL", "Tableau"], 31, "Save the Dolphins")

print("Name:             " , if_fellow.fname, if_fellow.lname)
print("Track:            ", if_fellow.track)
print("Cohort:           ", if_fellow.cohort_number)
print("Non-Profit Partner: ", if_fellow.nonprofit_partner)
print("Tools:            ", if_fellow.tools)
```

```python
class TFFellow(Fellow):

    def __init__(self, fname, lname, track, tools, corp_partner):

        # Step 1: Run Fellow's __init__ to set fname, lname, track, and tools
        super().__init__(fname, lname, track, tools)

        # Step 2: Set the attributes that are unique to TFFellow
        self.corp_partner   = corp_partner


# --- Test it out ---
tf_fellow = TFFellow("Alicia", "Torres", "Education Technology", ["Python", "HTML"], "Synchrony")

print("Name:         " , tf_fellow.fname, tf_fellow.lname)
print("Track:        " , tf_fellow.track)
print("Tools:        ", tf_fellow.tools)
print("Corp Partner: ", tf_fellow.corp_partner)
```

## 3. Polymorphism — Same Interface, Different Behavior

### What Is Polymorphism?

**Polymorphism** is the ability for different classes to define the same method name, each with its own unique behavior.

The word comes from Greek: *poly* (many) + *morph* (form). In object-oriented programming, it means that a single method name can take on **many forms** depending on which class is using it.

You have already seen this in action: every subclass of `Fellow` could define its own version of `switch_track()`. The name stays the same. The behavior can change per class. Code that calls `.switch_track()` does not need to care which type of fellow it is working with — it just calls the method and gets the right result.

---

### Why Polymorphism Matters

Consider a workforce platform that manages hundreds of fellows across multiple programs. At the end of each quarter, it needs to generate a status report for every fellow. With polymorphism, you can write one loop that calls `.generate_report()` on every fellow object — and each class handles the details of what that report contains.

Without polymorphism, you would need a separate code path for every single fellow type. Add a new program, rewrite the loop. Polymorphism eliminates that problem entirely.

**Key benefits:**
- One consistent interface across many different object types
- New classes can be added without changing existing code
- Reduces conditional logic ("if IFFellow do this, elif TFFellow do that...")

---

### Extending vs. Overriding

Polymorphism shows up in two ways:

| Pattern | What it means | When to use it |
|---|---|---|
| **Overriding** | Child class redefines a method from the parent | When the child needs different behavior for the same action |
| **Extending** | Child class adds brand-new methods the parent does not have | When the child has responsibilities the parent never had |

In the example below, `YouthProgramFellow` performs overriding for the `switch_track()` method.
---

```python
class YouthProgramFellow(Fellow):

    def __init__(self, fname, lname, track, tools, high_school, grad_year):

        # Call Fellow's __init__ to set the shared attributes
        super().__init__(fname, lname, track, tools)

        # Set the attributes unique to youth program fellows
        self.high_school    = high_school
        self.grad_year      = grad_year


    # OVERRIDING switch_track() from Fellow
    # Same method name, but this version adds a counselor approval notice.
    def switch_track(self, track):

        self.track = track
        print(
            self.fname + " " + self.lname +
            " has been moved to the " + self.track + " track." +
            " NOTE: A counselor must confirm this change within 5 business days."
        )
```

```python
youth_fellow = YouthProgramFellow(
    "Phil", "Harris", "Software Engineering",
    ["Python", "HTML", "CSS"],
    "Brooklyn Tech High School", 2026
)

print("Track: ", youth_fellow.track)
```

```python
# Demonstrating the overridden switch_track() — notice the counselor notice
youth_fellow.switch_track("Data Analytics")

print("Track: ", youth_fellow.track)
```
