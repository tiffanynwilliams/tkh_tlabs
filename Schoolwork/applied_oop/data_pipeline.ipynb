```python
from cleaner import clean_heartrate_data
import statistics as stats

class DataProcessor:
    """
    A class for loading and analyzing heart rate data from a file.
    """

    def __init__(self, filename: str):
        """
        Initialize the DataProcessor by storing the filename and
        reading its contents into self.data as a cleaned list.

        Args:
            filename (str): Path to the data file (e.g., 'data/phase0.txt').

        TODO:
            - Store the filename as an instance attribute
            - Open the file using file I/O and read lines into a list
            - Use cleaner.clean_heartrate_data() to clean the data
            - Store the cleaned result as self.data
    
        """
        # so the code is taking the DataProcessor class. we initialise the class at the very top. we use __init_ to do so.
        # the se;f os 
        self.filename = filename
        with open(self.filename) as f:
            lines = f.readlines()
        self.data = clean_heartrate_data(lines)
            

        
    def average(self) -> float:
        """
        Calculate and return the average of self.data.
        """
        return stats.mean(self.data)

    def median(self) -> float:
        """
        Calculate and return the median of self.data.
        """
        return stats.median(self.data)


    def range(self) -> float:
        """
        Calculate and return the range (max - min) of self.data.
        """
        return max(self.data) - min(self.data)
        pass
```
