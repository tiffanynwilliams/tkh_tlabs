## 1. What is Pandas — and Why Does It Matter?

**Pandas** is a Python library built for working with structured, tabular data — the kind of data you'd normally see in a spreadsheet or a database table. It was created in 2008 by Wes McKinney while he was working in finance, and it has since become the **default tool for data manipulation** across virtually every field that works with data.

### Why is Pandas so widespread?

| Field | How Pandas is Used |
|---|---|
| Finance | Analyzing stock prices, returns, risk calculations |
| Healthcare | Patient records, clinical trial data, outcomes analysis |
| Marketing | Customer segmentation, campaign performance, A/B testing |
| Sports Analytics | Player stats, game logs, performance trends |
| Machine Learning | Cleaning and preparing datasets before modeling |
| Journalism | Investigating public datasets, census data, government records |

The reason Pandas is everywhere is simple: **almost all real-world data is tabular**, and Pandas gives you a powerful, expressive toolkit to load, clean, reshape, filter, and summarize that data — all in a few lines of Python.

## The Two Core Data Structures

Before using Pandas, you need to understand its two fundamental building blocks:

### 1. `Series` — A single column of data
A **Series** is a one-dimensional labeled array. Think of it as a single column from a spreadsheet, with an index (row labels) on the left and values on the right.

### 2. `DataFrame` — A full table of data
A **DataFrame** is a two-dimensional table made up of multiple Series sharing the same index. Think of it as a full spreadsheet — rows and columns, each column being a Series.

````python
import pandas as pd

# ─────────────────────────────────────────────
# PART 1: Series
# A Series is a single column of labeled data.
# ─────────────────────────────────────────────

# Creating a Series from a plain Python list
temperatures = pd.Series([72, 68, 75, 80, 65], name="Temperature (°F)")
print("=== Series ===")
print(temperatures)
````

````python
# ─────────────────────────────────────────────
# PART 2: DataFrame
# A DataFrame is a table — a collection of
# Series that share the same index.
# ─────────────────────────────────────────────

# Creating a DataFrame from a dictionary
# Each key becomes a column name; each list becomes that column's values
data = {
    "Name":       ["Alice", "Bob", "Carmen", "David", "Eva"],
    "Age":        [28, 34, 22, 45, 30],
    "Department": ["Engineering", "Marketing", "Engineering", "HR", "Marketing"],
    "Salary":     [95000, 72000, 88000, 61000, 78000]
}

df = pd.DataFrame(data)
print("=== DataFrame ===")
print(df)
````

### 2.b Common Pandas Methods

#### The Pattern: Object . Method ( )

This should look familiar — it is the same pattern from **Object Oriented Programming**.

A DataFrame is an **object**. It has data stored inside it, and it comes with built-in **methods** that let you interact with that data. You call those methods using dot notation:

```
dataframe_name . method_name( )
       ↑                ↑
   the object       the action
```

For example:
```python
rent_df.head()      # call the head method on the rent_df object
rent_df.describe()  # call the describe method on the rent_df object
```

You already used this in regular Python — for example, `my_list.append(5)` or `my_string.upper()`. Pandas DataFrames follow the exact same idea. The object just happens to be a table of data.

---

### Methods We Will Cover

| Method | What It Does |
|---|---|
| `df.head()` | Shows the first N rows of the DataFrame (default is 5) |
| `df.describe()` | Prints summary statistics for all numeric columns |
| `df.info()` | Prints column names, data types, and non-null counts |
| `df.dropna()` | Returns a new DataFrame with all rows containing missing values removed |

````python
# ── df.head() ──────────────────────────────────────────────────────────────
# Shows the first rows of the DataFrame.
# By default it shows 5 rows, but you can pass a number to change that.

rent_df.head()
````

````python
rent_df.head(10)
````

````python
# ── df.info() ──────────────────────────────────────────────────────────────
# Prints a compact overview of the whole DataFrame:
#   - total number of rows
#   - each column's name
#   - how many non-null (non-missing) values it has
#   - the data type of each column  (int, float, object, etc.)
rent_df.info()
````

### 2.c Method Chaining

**Method chaining** is when you call one method directly on the result of another method — all in a single line.

The pattern looks like this:

```
dataframe . first_method( ) . second_method( )
```

This works because the first method returns an object, and that object immediately has the second method called on it. You never need to store the in-between result in a variable.

You have already seen this in regular Python:

```python
"  hello world  ".strip().upper()
```

Here, `.strip()` removes the whitespace and returns a new string, then `.upper()` is called on that result. Pandas works the exact same way.

---

### Example: `.isna().sum()`

- **`.isna()`** — goes through every cell in the DataFrame and returns `True` if the value is missing (`NaN`), and `False` if it is not
- **`.sum()`** — adds up all the `True` values column by column (remember: `True` counts as `1`, `False` as `0`)

Chained together, `rent_df.isna().sum()` gives you a count of missing values for every column in one step.

*(Note: student-authored code follows below.)*

````python
# Step 1 alone: .isna() returns a DataFrame of True/False values
rent_df.isna()
````

````python
# Step 2: chain .sum() onto .isna() to count the missing values per column
rent_df.isna().sum()
````

## 3. Cleaning Data in Pandas

Raw data is almost never perfect. Columns have missing values. Rows contain errors. Some information is simply not relevant to the question you are trying to answer. **Data cleaning** is the process of fixing or removing those problems before you do any analysis.

Pandas gives you several methods to handle this. The four most common are:

| Method | What It Does |
|---|---|
| `df.fillna(value)` | Fills missing values with whatever you pass in — a number, a string, or a calculated value like the column mean |
| `df.dropna()` | Removes any row that contains at least one missing value |
| `df.drop(...)` | Removes entire rows or columns that you no longer need |
| `df.replace(old, new)` | Swaps one value for another across the DataFrame |

---

### Think Before You Clean

These methods change the shape and content of your data, so **the decision of which one to use matters**.

Before calling any of these, ask yourself:

- **Why is this value missing?** Was it never collected, or does the absence itself mean something?
- **How much data would I lose** if I drop all rows with missing values?
- **Would filling a missing value with the mean make sense** for this column, or would it distort the data?
- **Is this column actually useful** for what I am trying to answer, or is it noise?

There is no single right answer. Choosing how to clean your data is an **executive decision** — it shapes every result that comes after it, so it deserves careful thought.

````python
# rent_df is already loaded — no import or read_csv needed


# ── df.fillna() ────────────────────────────────────────────────────────────
# Fill every missing value in the dataset with 0.
# Notice this returns a new DataFrame — rent_df itself is unchanged.

rent_df.fillna(0)
````

````python
# ── df.dropna() ────────────────────────────────────────────────────────────
# Remove every row that has at least one missing value.
# Compare the shape before and after to see how many rows were dropped.

print("Rows before dropna:", rent_df.shape[0])
print("Rows after  dropna:", rent_df.dropna().shape[0])
````

````python
# ── df.drop() ──────────────────────────────────────────────────────────────
# Remove columns you do not need.
# Here we drop the FIPS code and the HUD area code — useful for HUD internally,
# but not relevant if we are only analyzing rent prices.
# axis=1 tells Pandas we are dropping columns (axis=0 would drop rows).

rent_df.drop(columns=["fips", "hud_area_code"])
````

````python
# ── df.replace() ───────────────────────────────────────────────────────────
# Swap one value for another throughout the DataFrame.
# The metro column uses 1 and 0 — replace those with readable labels
# so the data is easier to interpret at a glance.

rent_df.replace({1: "Metro", 0: "Non-Metro"})
````

### 3.a Modification In Place

#### Your Changes Were Not Saved

Look back at every cleaning method we just called — `fillna`, `dropna`, `drop`, `replace`. Each one produced a result and Jupyter displayed it, but if you were to look at `rent_df` right now, it would be completely unchanged.

That is because by default, **Pandas methods return a new object — they do not modify the original**. The original DataFrame just sits there, untouched, until you explicitly tell Pandas to do otherwise.

---

### Two Ways to Make Your Changes Stick

#### Option 1 — Assign the result to a variable (recommended)

```python
clean_df = rent_df.dropna()
```

You are taking the result that Pandas hands back and storing it under a new name. From that point on, `clean_df` is your working copy and `rent_df` is still the original.

#### Option 2 — Use the `inplace=True` parameter

```python
rent_df.dropna(inplace=True)
```

This tells Pandas to apply the change directly to `rent_df` itself, overwriting it. No new variable is created.

---

### Why Option 1 Is Almost Always Better

Assigning to a new variable gives you two things at once:

1. **You keep the original.** If you make a mistake downstream, you can always go back to `rent_df` without reloading the file.
2. **The name tells a story.** A variable called `clean_df` or `metro_df` immediately signals to anyone reading your code that this is a modified version of the data — not the raw original.

Using `inplace=True` destroys the original without a safety net. If something goes wrong later, you have no fallback.

---

### When `inplace=True` Is Appropriate

There are situations where overwriting makes sense:

- You are working with a **very large dataset** and cannot afford to keep two copies of it in memory
- You are confident the transformation is correct and the original is already saved to disk (e.g., the CSV file is not going anywhere)
- You are at the **final step** of a cleaning pipeline and the intermediate versions have no further value

````python
# ── Proof that rent_df is unchanged ────────────────────────────────────────
# Even after all the cleaning methods above, rent_df still has its original shape.

print("rent_df shape (should be unchanged):", rent_df.shape)
print("Missing values still present:\n")

print(rent_df.isna().sum())
````

````python
# ── Option 1: Assign to a new variable (recommended) ──────────────────────
# Drop missing rows and store the result under a new name.
# rent_df remains the original — clean_df is our working copy.

clean_df = rent_df.dropna()

print("rent_df rows  (original):", len(rent_df))
print("clean_df rows (modified):", len(clean_df))
````

````python
# ── Option 2: inplace=True ─────────────────────────────────────────────────
# Here we work on a copy so we do not destroy rent_df for the rest of the notebook.
# In a real scenario where you intentionally want to overwrite, you would call
# this directly on your DataFrame — no copy needed.

demo_df = rent_df.copy()

print("demo_df rows before dropna:", len(demo_df))

demo_df.dropna(inplace=True)

print("demo_df rows after  dropna:", len(demo_df))
print()
print("rent_df is still untouched:", len(rent_df), "rows")
````

## 4. Indexing / Filtering Data

Once your data is loaded, you rarely want to work with the entire DataFrame at once. You will usually need to **pull out specific columns or rows** — that is what indexing is for.

There are three core ways to index into a Pandas DataFrame:

---

### 4.1 — Select a Single Column

Pass one column name inside single brackets. This returns a **Series**.

```python
rent_df["fmr_2"]
```

---

### 4.2 — Select Multiple Columns

Pass a **list** of column names inside the brackets. This returns a **DataFrame**.

```python
rent_df[["stusps", "countyname", "fmr_2"]]
```

The double brackets are not a typo — the outer brackets are the indexing operator, the inner brackets are the Python list.

---

### 4.3 — Select by Row and Column Using `.loc`

`.loc` lets you target a specific **row label** and **column name** at the same time.

```python
rent_df.loc[row_label, "column_name"]
```

By default, Pandas uses a numeric index (0, 1, 2, …) as row labels, so you pass the row number. You can also use a slice to grab a range of rows.

````python
# ── 4.1 Single column → returns a Series ──────────────────────────────────
rent_df["fmr_2"]
````

````python
# ── 4.2 Multiple columns → returns a DataFrame ────────────────────────────
rent_df[["stusps", "countyname", "fmr_1", "fmr_2", "fmr_3"]]
````

````python
# ── 4.3 .loc — target a specific row and column ───────────────────────────

# A single cell: row 0, column "hud_area_name"
print("Single cell with .loc:")
print(rent_df.loc[0, "hud_area_name"])
````

````python
# A single full row
print("Full row 0:")
print(rent_df.loc[0])
````

````python
# A slice of rows, specific columns
print("Rows 0 through 4, selected columns:")
rent_df.loc[0:4, ["stusps", "countyname", "fmr_2"]]
````

## 4.a Boolean Filters

### What Is a Boolean Filter?

A **boolean filter** is a condition you write that evaluates every row in the DataFrame and produces an array of `True` and `False` values — one result per row.

For example:

```python
rent_df["fmr_2"] > 2000
```

This does not return a number. It returns a **Series of True/False values** — `True` for every row where the 2-bedroom rent is above $2,000, and `False` for every row where it is not.

---

### Step 1 — Create the boolean expression

```python
mask = rent_df["fmr_2"] > 2000
```

`mask` is now a Series that looks like this:

```
0      False
1      False
2       True
3      False
...
```

---

### Step 2 — Plug the mask into the DataFrame brackets

```python
rent_df[mask]
```

Pandas goes through every row. If the mask value for that row is `True`, the row is kept. If it is `False`, the row is dropped. The result is a filtered DataFrame containing **only the rows where your condition was met**.

You can also write it in one line without the intermediate variable:

```python
rent_df[rent_df["fmr_2"] > 2000]
```

Both approaches produce the exact same result — the two-step version with a named mask is simply easier to read.

````python
# ── Step 1: See the boolean array on its own ──────────────────────────────
# Every row gets a True or False depending on the condition.

mask = rent_df["fmr_2"] > 2000

mask
````

````python
# ── Step 2: Plug the mask into the DataFrame ──────────────────────────────
# Only rows where mask == True are returned.

rent_df[mask]
````

````python
# ── One-liner version — same result, no intermediate variable ──────────────
rent_df[rent_df["fmr_2"] > 2000]
````
