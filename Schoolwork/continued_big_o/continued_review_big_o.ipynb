## 1. What is Time Complexity?

**Time complexity** is a way of describing how the runtime of a program grows as the size of its input grows. Instead of measuring exact seconds, we describe the *pattern* of growth — does the program get a little slower with more data, or dramatically slower?

---

## Key Definitions

**Time Complexity**
A measure of how the runtime of an algorithm grows as the size of its input grows. We don't measure exact seconds — we describe the *pattern* of growth so we can compare algorithms fairly regardless of hardware.

**Worst-Case Time Complexity**
The maximum amount of time an algorithm could ever take for an input of size *n*. We use worst-case because it gives us a guarantee — no matter what data you throw at the algorithm, it will never perform worse than this bound. Big O notation always describes worst-case.

**Constant Time — O(1)**
The algorithm takes the same amount of time no matter how large the input is. Accessing a single element in a list by index is O(1) — it doesn't matter if the list has 10 items or 10 million.

**Logarithmic Time — O(log(n))**
The runtime whose steps grows in proportion the logarithm of the input size. One of the most efficient time complexities because the number of operations increase very slowly even as input size grows.

**Linear Time — O(n)**
The runtime grows proportionally with the input size. If the input doubles, the runtime roughly doubles. Looping through every item in a list once is O(n).

**Quadratic Time — O(n²)**
The runtime grows proportionally to the *square* of the input size. If the input doubles, the runtime roughly quadruples. A loop inside a loop (where both depend on *n*) is the classic example of O(n²).

**Cubic Time — O(n³)**
The runtime grows proportionally to the *cube* of the input size . A loop inside a loop inside another loop (where both depend on *n*) is the classic example of O(n³).

## 2. Linear Runtime — O(n)

When the number of steps always equals the size of the list, we state that the Big O notation of algorithm is linear (aka O(n)). 

A list of 1 takes 1 step. A list of 10 takes 10 steps. A list of 100 takes 100 steps.

### How to spot O(n)

Look for a single loop that travels through the input once. Every element gets visited exactly one time, so the work grows at the same rate as the input.

```python
for num in nums:   # runs once for every element → O(n)
    ...
```

### Calculating the true runtime

Not every linear algorithm does exactly one unit of work per element. Sometimes each step involves a few operations. We can count them:

```python
def find_steps(nums, x):
    steps = 0        # 1 operation (happens once)
    for num in nums:
        steps += 1   # 1 operation per element
        if num == x: # 1 operation per element
            return steps
    return steps     # 1 operation (happens once)
```

Inside the loop there are **2 operations** per element, and **2 operations** that happen once regardless of input size. So the true runtime is:

> **2n + 2**

For a 10-element list: `2(10) + 2 = 22` operations.  
For a 100-element list: `2(100) + 2 = 202` operations.

### Dropping the constants

When *n* gets very large, the constants become insignificant. Whether it is `2n + 2`, `4n + 5`, or `3n`, they all grow at the same *rate* — one straight line. So we simplify:

> `2n + 2` → **O(n)**  
> `4n + 5` → **O(n)**  
> `3n` → **O(n)**

We drop the coefficients and the added constants and keep only the dominant term. The label **O(n)** tells us the shape of the growth, not the exact count. That is all we need to compare algorithms.

````python
def find_steps(nums, x):
    """
    Count up how many steps this algorithm takes to complete.
    """
    steps = 0
    for num in nums:
        steps += 1
        if num == x:
            return steps
    return steps
````

## 3. Deceptively Linear Built-in Operations — Still O(n)

Some Python built-in operations look instant because they fit on a single line and have no visible loop. But looks can be deceiving — under the hood, Python still has to visit every element in the collection to compute the answer.

### Why does this happen?

Python lists are stored as sequences in memory. There is no shortcut to answer questions like *"is 7 in this list?"* or *"what is the largest value?"* without checking each element one by one. The loop is just hidden inside the function — you don't write it, but the CPU still executes it.

### Common examples

| Operation | Looks like | Actually |
|---|---|---|
| `x in my_list` | One check | Scans every element until found — O(n) |
| `max(my_list)` / `min(my_list)` | One call | Walks the entire list to compare — O(n) |
| `sum(my_list)` | One call | Adds every element one by one — O(n) |
| `my_list.count(x)` | One call | Scans every element to count matches — O(n) |
| `my_list.index(x)` | One call | Scans from the front until found — O(n) |

### The key insight

The number of operations you *write* has nothing to do with the time complexity. What matters is how many operations the CPU *runs* — and for all of the above, that number grows linearly with the size of the list.

### Example 1: The `in` operator

`x in my_list` reads like a single check, but Python has to scan from the beginning of the list and compare each element until it either finds a match or runs out of elements.

The worst case — when the value is **not** in the list — forces Python to inspect every single element before it can say "no". That is O(n).

Below we write our own `manual_in` that makes the hidden loop explicit, then confirm it takes exactly as many steps as there are elements in the list.

````python
def linear_in(nums, x):
    """
    Does exactly what `x in nums` does — exposes the hidden loop.
    Returns (found, steps) so we can see the work being done.
    """
    steps = 0
    for num in nums:
        steps += 1          # one comparison per element
        if num == x:
            return True, steps
    return False, steps
````

````python
# Search for -1 (guaranteed not present) so we always hit worst case
found_1,   steps_1   = manual_in(list_1,   -1)
found_10,  steps_10  = manual_in(list_10,  -1)
found_100, steps_100 = manual_in(list_100, -1)

print(found_1, "Steps taken", steps_1)
print(found_10, "Steps taken", steps_10)
print(found_100, "Steps taken", steps_100)
````

### Example 2: `max()` and `min()`

`max(my_list)` feels like a single lookup, but Python has no magic way to know which element is largest without comparing every element to the current best candidate — one by one, from left to right.

The manual version below reveals the loop:

````python
def manual_max(nums):
    """
    Does exactly what max(nums) does — exposes the hidden loop.
    Returns (maximum, steps).
    """
    steps = 0
    best = nums[0]          # assume the first element is the largest
    for num in nums:
        steps += 1          # one comparison per element
        if num > best:
            best = num
    return best, steps
````

````python
max_1,   steps_1   = manual_max(list_1)
max_10,  steps_10  = manual_max(list_10)
max_100, steps_100 = manual_max(list_100)

print(max_1, "Steps taken", steps_1)
print(max_10, "Steps taken", steps_10)
print(max_100, "Steps taken", steps_100)
````

### Example 3: `list.count()`

`my_list.count(x)` returns how many times `x` appears in the list. To do this, Python **must** scan the entire list every time — it cannot stop early, because a match might appear anywhere, including the very last position.

This makes `.count()` always O(n), even if the value is not present at all.

````python
def manual_count(nums, x):
    """
    Does exactly what nums.count(x) does — exposes the hidden loop.
    Returns (count, steps).
    """
    steps = 0
    total = 0
    for num in nums:
        steps += 1          # one comparison per element — no early exit possible
        if num == x:
            total += 1
    return total, steps
````

````python
# Count how many times -1 appears (it doesn't — but the full scan still happens)
count_1,   steps_1   = manual_count(list_1,   -1)
count_10,  steps_10  = manual_count(list_10,  -1)
count_100, steps_100 = manual_count(list_100, -1)

print(count_1, "Steps taken", steps_1) 
print(count_10, "Steps taken", steps_10)
print(count_100, "Steps taken", steps_100)
````

## 4. Quadratic Time — O(n²)

When an algorithm is **quadratic**, its runtime grows proportionally to the *square* of the input size.

- A list of 2 → ~4 steps  
- A list of 10 → ~100 steps  
- A list of 100 → ~10,000 steps  

Double the input, and the runtime **quadruples**. The growth is not a straight line — it curves upward steeply.

### How to spot O(n²)

The signature pattern is a **loop inside a loop**, where both loops depend on the size of the same input:

```python
for i in nums:        # runs n times
    for j in nums:    # runs n times for every i
        ...           # this code runs n × n = n² times total
```

For every step the outer loop takes, the inner loop takes *n* steps of its own. The total work is n × n = **n²**.

### Why this matters

Quadratic algorithms can handle small inputs just fine, but they slow down dramatically as the input grows. An algorithm that handles 100 items in a fraction of a second may take *minutes* on 10,000 items. Recognizing the nested-loop pattern is the first step to knowing when to look for a faster approach.

````python
def count_pairs(nums):
    """
    Finds every possible pair of elements in the list.
    Uses a loop inside a loop — a classic O(n²) pattern.
    """

    steps = 0

    for i in nums:          # outer loop — runs once per element
        for j in nums:      # inner loop — runs once per element, for every outer step
            steps += 1      # one step each time the inner body executes

    return steps
````

````python
steps_1   = count_pairs(list_1)
steps_10  = count_pairs(list_10)
steps_100 = count_pairs(list_100)

print("Steps taken", steps_1)
print("Steps taken", steps_10)
print("Steps taken", steps_100)
````

## 5. Making a Faster Search

### Revisiting Linear Search

We have already seen how searching a list works. To find a value, Python starts at the very first element and checks each one in order — left to right — until it either finds a match or runs out of elements to check.

```python
for num in nums:
    if num == target:
        return True
return False
```

This is **linear search** — O(n). For a list of 10 elements that is perfectly fine. But imagine searching a list of **1,000,000** elements. In the worst case — when the value is not in the list at all — we would make one million comparisons just to come back empty handed.

---

### How can we make this faster?

Let's think about this like a real world problem.

Imagine you are looking up the name **"Smith"** in a phone book. Would you start on page 1 and read every single name in order until you found it? Of course not. You would flip open somewhere in the middle, see that you landed in the M's, and immediately know that Smith must be further right. You jump again, narrow it down, and find the name in a handful of flips.

That instinct is the key idea. You were able to skip huge sections of the book because it was **sorted**. You never had to read everything — you used the order of the data to your advantage.

---

### First thought: what if we sort the list first?

If a list is sorted, we immediately know something powerful about every single element:

- Everything to the **left** of any position is *smaller*
- Everything to the **right** of any position is *larger*

This means the moment we check a value and our target is smaller than it, we can completely ignore everything to the right. We just eliminated half the list in a single step — without looking at a single element in that half.

That one idea — *eliminate half the remaining elements with every comparison* — is the intuition behind one of the most important algorithms in computer science. We will build it in the next section.

### Searching a Sorted List — Step by Step

Let's put this idea into practice. We have a sorted list and we want to find the number **64**.

```
[10, 13, 28, 29, 40, 43, 64, 69]
```

Instead of starting from the left and checking every number one by one, we are going to use the same instinct as the phone book — **open to the middle and use what we see to throw away half the list**.

---

**Step 1 — Open to the middle**

The list has 8 elements. The middle element is at index 3, which holds the value **29**.

```
[10, 13, 28, 29, 40, 43, 64, 69]
              ^^
           middle = 29
```

We are looking for 64. Is 64 greater than 29? **Yes.**

That means 64 cannot be anywhere in the left half. We close that half entirely and never look at it again.

```
[10, 13, 28, 29, | 40, 43, 64, 69]
  ignore this    |  search here
```

In one step we went from 8 candidates down to **4**.

---

**Step 2 — Open to the middle of what remains**

Our remaining list is `[40, 43, 64, 69]`. The middle element is **43**.

```
[40, 43, 64, 69]
     ^^
  middle = 43
```

We are looking for 64. Is 64 greater than 43? **Yes.**

Once again, we close the left half and ignore it.

```
[40, 43, | 64, 69]
 ignore  |  search here
```

We went from 4 candidates down to **2**.

---

**Step 3 — One more flip**

Our remaining list is `[64, 69]`. The middle element is **64**.

```
[64, 69]
 ^^
middle = 64
```

Is 64 equal to 64? **Yes — found it!**

---

### What just happened?

Linear search would have checked: 10, 13, 28, 29, 40, 43, 64 — that is **7 comparisons**.

Using the phone book approach, we found 64 in just **3 comparisons**.

And crucially, each step did not just eliminate one element — it eliminated **half of everything that was left**. That is what makes this approach so powerful. The bigger the list, the more dramatic the difference becomes.

````python
def binary_search(nums, target):
    begin = 0
    end = len(nums) - 1
    mid = 0

    steps = 0

    # TODO: which conditional can we use to continue checking the list until the beginning
    # surpasses the end?
    # keep looping while bgein is not more than end
    while begin < end:
        steps += 1
        # TODO: how can we calculate "mid" to be the middle number between high & low?
        mid = (begin + end) // 2

        # TODO: what should we do if the middle number is less than our target?
        if nums[mid] < target:
            begin = mid + 1
        # TODO: what if greater than?
        elif nums[mid] > target:
            end = mid - 1
        # nums[mid] == target, return mid & target
        else:
            return mid, steps
    
    return -1, steps
````

````python
numbers  = [10, 13, 28, 29, 40, 43, 64, 69]

index, steps = binary_search(numbers, 64)

print("Found 64 at position", index, "Steps taken is", steps)
````

#### The two big traps to remember

**Trap 1 — List vs dictionary `in`**
`x in my_list` is O(n). `x in my_dict` is O(1). The same keyword does very different things depending on the data structure.

**Trap 2 — Built-ins called inside loops**
`sum()`, `max()`, `.count()`, and `.index()` are each O(n) on their own. Calling any of them inside a loop turns that loop into O(n²).
